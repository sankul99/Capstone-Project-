```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
df_crimes = pd.read_csv(r'C:\Users\sanku\OneDrive\Desktop\India_Crime - Copy\India_Crime - Copy\01_District_wise_crimes_committed_IPC_2001_2012.csv')
```

```python
df_crimes.head()
```

```python
main_df = df_crimes[['STATE/UT', 'DISTRICT', 'YEAR',  'TOTAL IPC CRIMES']]
```

```python
main_df.head()
```

```python
new_df = df_crimes.iloc[:,3:]
```

```python
new_df.head()
```

```python
scaler = MinMaxScaler()
data_scaled = scaler.fit_transform(new_df)
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++')
kmeans.fit(data_scaled)
```

```python
df_crimes['cluster'] = kmeans.labels_
```

```python
peaceful_areas = df_crimes[df_crimes['cluster'] == 0]
sensitive_areas = df_crimes[df_crimes['cluster'] == 1]
moderate_areas = df_crimes[df_crimes['cluster'] == 2]
```
